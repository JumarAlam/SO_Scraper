```python
import csv
import pandas as pd
import numpy as np
```

```python
dfQuestions = pd.read_csv('Questions.csv', delimiter="|", names=["qID", "qHeader", "qvote", "question", "qCodePart", "qPostTime", "qAuthor", "qAuthorRep", "numAnswers"])
```

```python
dfQuestions
```

```python
# Number of Unique Questions
len(dfQuestions['qID'].unique())
```

```python
dfQuestions = dfQuestions.drop_duplicates("qID", ignore_index=True)
```

```python
dfQuestions
```

```python
# Converting all [] into NA in qCodePart column
dfQuestions = dfQuestions.replace('[]', pd.NA)
dfQuestions
```

```python
# Remove all the rows where Code Part is empty

dfQuestions = dfQuestions.dropna()

dfQuestions.reset_index(drop = True, inplace = True)
len(dfQuestions)
```

```python
dfQuestions
```

```python
dfQuestions.to_csv("CleanedQuestions.csv", sep = "|",  index=False)
```
