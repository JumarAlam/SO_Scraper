```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generating a graph for Number of Post per year
```

```python
QnAMerged = pd.read_csv('CleanednMerged.csv', delimiter="|")
```

```python
dateModDf = QnAMerged
```

```python
dateModDf['qPostTime'] = dateModDf['qPostTime'].str[:-9]
```

```python
dateModDf['qPostTime'] = pd.to_datetime(dateModDf['qPostTime'])
dateModDf['qPostTime']
```

```python
sortByDate = dateModDf.sort_values('qPostTime').set_index('qPostTime')
sortByDate
```

```python
yearly_group = sortByDate.resample('Y').count()
```

```python
yearly_group
```

```python
# ax = monthly_group.plot(kind='bar',
#                                figsize=(10,5),legend=None)
# ax.set_xticklabels(sortByDate.index.strftime('%Y-%m-%d'), rotation=90)
# ax.set_yticklabels(sortByDate['qID'])
# plt.xlabel('Date - Month Starting')

fig = plt.figure(figsize = (10, 5))
#bars = plt.bar(monthly_group['qID'], monthly_group['qID'])
#bars = [25,0,0,5,0,14,144,305,645,1532,2805,4119,4757,4494]
plt.bar(yearly_group.index.strftime('%Y'), yearly_group['qID'])

plt.xlabel('Year')
plt.ylabel('Number of Posts Each Year')
plt.show()
```

```python
# Generating Graph based on number of votes
```

```python
#QnAMerged
sortbyVotes = QnAMerged.sort_values('qvote')
sortbyVotes
```

```python
BelowZero = 0 
for i in range(len(sortbyVotes)):
    if sortbyVotes['qvote'][i] < 0:
        BelowZero += 1
BelowZero
```

```python
EqualsZero = 0 
for i in range(len(sortbyVotes)):
    if sortbyVotes['qvote'][i] == 0:
        EqualsZero += 1
EqualsZero
```

```python
BetweenZeroNTen = 0 
for i in range(len(sortbyVotes)):
    if (sortbyVotes['qvote'][i] > 0) & (sortbyVotes['qvote'][i] <= 10):
        BetweenZeroNTen += 1
BetweenZeroNTen
```

```python
Above10 = 0 
for i in range(len(sortbyVotes)):
    if sortbyVotes['qvote'][i] > 10 :
        Above10 += 1
Above10
```

```python
vote_group = sortbyVotes.groupby(['qvote']).count()
vote_group
```

```python
fig = plt.figure(figsize = (10, 5))
#bars = plt.bar(monthly_group['qID'], monthly_group['qID'])
#bars = [25,0,0,5,0,14,144,305,645,1532,2805,4119,4757,4494]
plt.bar(vote_group.index, vote_group['qID'])

plt.xlabel('Vote Count')
plt.ylabel('Number of Posts Based on Vote Count')
plt.show()
```
